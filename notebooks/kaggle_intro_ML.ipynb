```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor

pd.options.display.float_format = '{:.0f}'.format
```

```python
data_path = "../data/melb_data.csv"

house_data = pd.read_csv(data_path)
```

```python
house_data.describe()
```

```python
# removendo os NAs
clean_house_data = house_data.dropna(axis=0)
```

```python
# analisar as colunas da relação
clean_house_data.columns
```

```python
# cria o y que queremos realizar a predição, no caso preço
y = clean_house_data.Price

# separa as colunas da base original que vão caracterizar os imóveis e serão usados para predição
features = [ 'Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude' ]

# cria a base de dados
X = clean_house_data[features]
```

# Decision Tree models

```python
# cria o modelo de árvore de decisão
model_tree_in_sample = DecisionTreeRegressor(random_state=42)

# in-sample predictions
model_tree_in_sample.fit(X, y)

# predição usando base inteira
in_sample_val = model_tree_in_sample.predict(X)

# analisar alguns dados preditos com relação ao y
print(in_sample_val[:5])
print(y.head(5).to_list())
```

```python
# média absoluta de erro nos dados de validação 
val_mea = mean_absolute_error(y, in_sample_val)

print(val_mea)
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# comparar MAE com diferentes valores de max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    local_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(f"Max leaf nodes: {max_leaf_nodes}  \t\t Mean Absolute Error:  {local_mae}")
```
